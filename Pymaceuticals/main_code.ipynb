```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_file = "../Instructions/Pymaceuticals/data/mouse_drug_data.csv"
clinical_trial_file = "../Instructions/Pymaceuticals/data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
df_mouse = pd.read_csv(mouse_file)
df_clinical = pd.read_csv(clinical_trial_file)

# Combine the data into a single dataset
df_combine = df_clinical.merge(df_mouse, on = "Mouse ID", how = "inner")
df_combine = df_combine.sort_values(by=["Timepoint", "Tumor Volume (mm3)"])
# Display the data table for preview
df_combine.head()
```

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
gb = df_combine.groupby(["Drug", "Timepoint"])

# Convert to DataFrame
mean_tumor_df = gb.mean()

# Preview DataFrame
mean_tumor_df = mean_tumor_df.reset_index()
mean_tumor_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
gb = df_combine.groupby(["Drug", "Timepoint"])
# Convert to DataFrame
sem_tumor_df = gb.sem()
sem_tumor_df = sem_tumor_df.reset_index()

# Preview DataFrame
sem_tumor_df = sem_tumor_df[["Drug", "Timepoint", "Tumor Volume (mm3)"]]
sem_tumor_df
```

```python
# Minor Data Munging to Re-Format the Data Frames
mean_tumor_pivot = mean_tumor_df.pivot(index = "Timepoint", columns = "Drug")
sem_tumor_pivot = sem_tumor_df.pivot(index = "Timepoint", columns = "Drug")
# Preview that Reformatting worked
mean_tumor_pivot.head()
```

```python
# Generate the Plot (with Error Bars)
times = []
for row in sem_tumor_pivot.index:
    times.append(row)
x_axis = times


plt.plot( x_axis, mean_tumor_pivot["Tumor Volume (mm3)"]["Capomulin"], marker='o', markerfacecolor='red', color='red', label = "Capomulin")
plt.plot( x_axis, mean_tumor_pivot["Tumor Volume (mm3)"]["Infubinol"], marker='^', markerfacecolor='blue', color='blue', label = "Infubinol")
plt.plot( x_axis, mean_tumor_pivot["Tumor Volume (mm3)"]["Ketapril"], marker='s', markerfacecolor='green', color='green', label = "Ketapril")
plt.plot( x_axis, mean_tumor_pivot["Tumor Volume (mm3)"]["Placebo"], marker='d', markerfacecolor='black', color='black', label = "Placebo")

plt.errorbar(x_axis, mean_tumor_pivot["Tumor Volume (mm3)"]["Capomulin"], yerr = sem_tumor_pivot["Tumor Volume (mm3)"]["Capomulin"], fmt = ' ', color = "red")
plt.errorbar(x_axis, mean_tumor_pivot["Tumor Volume (mm3)"]["Infubinol"], yerr = sem_tumor_pivot["Tumor Volume (mm3)"]["Infubinol"], fmt = ' ', color = "blue")
plt.errorbar(x_axis, mean_tumor_pivot["Tumor Volume (mm3)"]["Ketapril"], yerr = sem_tumor_pivot["Tumor Volume (mm3)"]["Ketapril"], fmt = ' ', color = "green")
plt.errorbar(x_axis, mean_tumor_pivot["Tumor Volume (mm3)"]["Placebo"], yerr = sem_tumor_pivot["Tumor Volume (mm3)"]["Placebo"], fmt = ' ', color = "black")

plt.legend(loc='upper left')
plt.rcParams["axes.grid.axis"] ="y"
plt.rcParams["axes.grid"] = True

plt.xlabel("Time (Days)")
plt.ylabel("tumor Volume (mm3)")
plt.title("Tumor Response to Treatment")

plt.savefig("Tumor_Respons_to_Treatment.png")

plt.show()
# Save the Figure

```

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mouse_survival = df_combine.groupby(["Drug", "Timepoint"]).count()
# Convert to DataFrame
mouse_survival = mouse_survival.reset_index()
mouse_survival = mouse_survival[["Drug", "Timepoint", "Mouse ID"]]
mouse_survival = mouse_survival.rename(columns = {"Mouse ID": "Mouse Count"})
# Preview DataFrame
mouse_survival.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mouse_survival_pivot = mouse_survival.pivot(index = "Timepoint", columns = "Drug")

# Preview the Data Frame
mouse_survival_pivot.head()
```

```python
# Generate the Plot (Accounting for percentages)
times = []
for row in sem_metastatic_pivot.index:
    times.append(row)
x_axis = times


plt.plot( x_axis, (mouse_survival_pivot["Mouse Count"]["Capomulin"]/25)*100, marker='o', markerfacecolor='red', color='red', label = "Capomulin")
plt.plot( x_axis, (mouse_survival_pivot["Mouse Count"]["Infubinol"]/25)*100, marker='^', markerfacecolor='blue', color='blue', label = "Infubinol")
plt.plot( x_axis, (mouse_survival_pivot["Mouse Count"]["Ketapril"]/25)*100, marker='s', markerfacecolor='green', color='green', label = "Ketapril")
plt.plot( x_axis, (mouse_survival_pivot["Mouse Count"]["Placebo"]/25)*100, marker='d', markerfacecolor='black', color='black', label = "Placebo")

plt.legend(loc='lower left')
plt.rcParams["axes.grid.axis"] ="y"
plt.rcParams["axes.grid"] = True

plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")
plt.title("Survival During Treatment")

plt.savefig("Survival_During_Treatment.png")

# Show the Figure
plt.show()
```

```python
# Calculate the percent changes for each drug
num_list = []
percent_changes = []
drugs = ["Capomulin", "Infubinol", "Ketapril", "Placebo"]
for column in mean_tumor_pivot["Tumor Volume (mm3)"]:
    num_list = []
    if column in drugs:
        for item in mean_tumor_pivot["Tumor Volume (mm3)"][column]:
            num_list.append(item)
        percent_change = ((num_list[-1]-num_list[0])/num_list[0]*100)
        percent_changes.append(percent_change)

percent_df = pd.DataFrame({"Drug": drugs,
                         "Percent Changes": percent_changes})

# Display the data to confirm
percent_df
```

```python
ypos = np.arange(len(percent_changes))

abovezero_list = []
for item in percent_changes:
    if item > 0:
        abovezero_list.append(True)
    else:
        abovezero_list.append(False)

percent_df["above"] = abovezero_list
    
plt.title("Tumor Change Over 45 Day Treatment")
plt.ylabel("% Tumor Volume Change")

formatted_percent = ['%.2f' % elem for elem in percent_changes]

for i, v in enumerate(formatted_percent):
    plt.text(ypos[i] - 0.25, 0, str(v), color = "black")

plt.xticks(ypos, drugs)
plt.bar(ypos, percent_changes, color = percent_df.above.map({True: "r", False: "g"}))

plt.savefig("Tumor_Change_Over_45_Day_Treatment.png")

plt.show()
```
